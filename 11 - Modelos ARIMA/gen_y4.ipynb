```python
import numpy as np
import pickle
from matplotlib import pyplot as plt
%matplotlib inline

from statsmodels.graphics.tsaplots import plot_predict
from statsmodels.tsa.arima_process import arma_generate_sample
from statsmodels.tsa.arima.model import ARIMA

from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf

import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(12345)

ma2 = [1, 0.9, 0.7, 0.5, 0.3, 0.1]
ar2 = np.convolve([1, -1.5, 0.7],[1, -1])
y4 = arma_generate_sample(ar2, ma2, 10000)
```

```python
plt.figure(figsize=[15, 4])
ax1 = plt.subplot(1,3,1)
ax1.set_xlabel('Samples')
ax1.set_ylabel('Time series')
ax1.plot(y4[:50],'mo--'), plt.grid()

ax2 = plt.subplot(1,3,2)
ax2.set_xlabel('Lag [Samples]')
ax2.set_ylabel('ACF')
plot_acf(y4,ax=ax2,lags=20)
ax2.grid()

ax3 = plt.subplot(1,3,3)
ax3.set_xlabel('Lag [Samples]')
ax3.set_ylabel('P-ACF')
plot_pacf(y4,ax=ax3,lags=20)
ax3.grid()
plt.show()
```
